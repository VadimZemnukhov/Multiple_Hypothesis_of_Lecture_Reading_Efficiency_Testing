Эти колонки можно убрать.

Нам необходимо проверить гипотезу о том, улучшается ли результат прохождения теста после прочнения лекции. Т.е. самый первый тест, который был на лекции, будет иметь самую раннюю дату. Любую другую дату прохождения теста можно считать датой прохождения после прочнения лекции. При этом нас не интересует время, а только дата, т.к. студент мог во время лекции в любой момент пройти тест (хоть это и не приветствовалось).

```python
df_4_1 = df_4_1.iloc[:, [1, 4, 6, 8, 14, 15]]
```

```python
df_4_1['Время создания'] = pd.to_datetime(df_4_1['Время создания']).dt.date
```

```python
df_4_1.head(5)
```

```python
first_date = df_4_1['Время создания'].min()
```

```python
X = np.array(df_4_1[df_4_1['Время создания'] == first_date]['Набрано баллов'])
```

```python
Y = np.array(df_4_1[df_4_1['Время создания'] != first_date]['Набрано баллов'])
```

```python
Y
```

```python
X
```

Проверим нормальность распределения при помощи нескольких критериев. Если все критерии будут говорить о нормальности, то будем считать эту информацию достоверной. Остальные гипотезы об улучшении результатов прохождения тестов будем считать по методу множественной проверки гипотез методом Холма.

```python
sps.normaltest(X)
```

```python
sps.normaltest(Y)
```

Гипотеза о нормальности не отвергается на уровне значимости 5%.

```python
sps.shapiro(X)
```

```python
sps.shapiro(Y)
```

Солгасно критерию Шапиро-Уилка гипотеза о нормальности не отвергается на уровне значимости 5%.

```python
sps.jarque_bera(X)
```

```python
sps.jarque_bera(Y)
```

Солгасно критерию Харке-Бэра гипотеза о нормальности не отвергается на уровне значимости 5%.

```python
from statsmodels.stats.diagnostic import lilliefors

lilliefors(X)
```

```python
lilliefors(Y)
```

Солгасно критерию Лилиефорса гипотеза о нормальности не отвергается на уровне значимости 5%.

Считаем информацию о нормальности распределения двух выборок достоверной.

**Нулевая гипотеза:** после прочтения лекции 4 результаты прохождения тестов по первой её половине не изменились. \
**Альтернативная гипотеза:** после прочтения лекции 4 результаты прохождения тестов по первой её половине стали выше.

Нам также нужно выполнить аналогичные проверки для тестов по второй половине 4-й лекции, а также по 5-й лекции. Итого 3 гипотезы. Если гипотезы будут отвергнуты, то можно будет сделать вывод, что в среднем после прочтения лекции результаты тестов становятся выше, если наоборот, то значит можно будет сделать вывод, что прочтение лекций не влияет на результат написания тестов. Если же в каких-то случаях мы получим отклонение нулевой гипотезы, а в каких-то "не отклонение", то однозначного вывода о картине в целом сделать будет нельзя, но можно будет с определенной уверенностью утверждать о результатах по конкретным тестам.

Для проверки первой гипотезы примем уровень значимости 5/3 = 1,67%.

Т.к. обе выборки из нормального распределения, имеют небольшой размер, при этом зависимы, проведем парный t-test.

```python
sps.ttest_ind(X, Y)
```

Нулевая гипотеза не отклоняется на уровне значимости 1,67%.
